##  第一题

```python
nums = input()
nums = int(nums)
result = [0]*nums
for i in range(nums):
    n = int(input())
    test = []
    for j in range(n):
        test.append(input().split())
    result[i]=test
    
for i in range(nums):
    x = 0
    y = 0
    k = 0
    while y<len(result[i]):
        k += int(result[i][y][x])
        y += 1
        x += 1

    
    r = 0
    c = 0
    N = len(result[i])
    for row in result[i]:
        if len(set(row))<N:
            r += 1
    for col in range(N):
        if len(set(q[col] for q in result[i]))<N:
            c += 1
    print("Case #"+str(i+1)+":",str(k),str(r),str(c))
```

## 第二题

```python
def right(s):
    end = 0
    for i in range(len(s)-1,-1,-1):
        if s[i] == ")":
            end += 1
        else:
            return end
def remove(s,n):
    return s[n:-n]

nums = int(input())
result = [0]*nums
for i in range(nums):
    result[i] = list(input())

for i in range(nums):
    for j in range(len(result[i])):
        result[i][j] = "("*int(result[i][j])+result[i][j]+")"*int(result[i][j])
    while len(result[i])>1:
        if result[i][0][-1] == '0' or result[i][1] == '0':
            result[i][0] += result[i][1]
            result[i].pop(1)
        else:
            mind = min(right(result[i][0]),right(result[i][1]))
            result[i][1] =remove(result[i][1],mind)
            result[i][0] = result[i][0][:-mind]+result[i][1]+result[i][0][-mind:]
            result[i].pop(1)
    print("Case #"+str(i+1)+":",result[i][0])
            
result
```

## 第三题

```python
nums = input()
nums = int(nums)
result = [0]*nums
for i in range(nums):
    n = int(input())
    test = []
    for j in range(n):
        temp = input().split()
        temp = [int(e) for e in temp]
        test.append(temp+[j])
    result[i]=test

for i in range(nums):
    J = 0
    G = 0
    F = 0
    result[i].sort(key=lambda k: k[0])
    length = len(result[i])
    for t in range(length):
        if result[i][t][0] >= J:
            J = result[i][t][1]
            result[i][t] += ["J"]
        elif result[i][t][0] >= G:
            G = result[i][t][1]
            result[i][t] += ["C"]
        else:
            print("Case #"+str(i+1)+":","IMPOSSIBLE")
            F = 1
            break
    if F == 0:
        result[i].sort(key=lambda k:k[2])
        answer = list(k[3] for k in result[i])
        answer = "".join(answer)
        print("Case #"+str(i+1)+":",answer)
result
```
