##  第一题

```python
nums = input()
nums = int(nums)
result = [0]*nums
for i in range(nums):
    n = int(input())
    test = []
    for j in range(n):
        test.append(input().split())
    result[i]=test
    
for i in range(nums):
    x = 0
    y = 0
    k = 0
    while y<len(result[i]):
        k += int(result[i][y][x])
        y += 1
        x += 1

    
    r = 0
    c = 0
    N = len(result[i])
    for row in result[i]:
        if len(set(row))<N:
            r += 1
    for col in range(N):
        if len(set(q[col] for q in result[i]))<N:
            c += 1
    print("Case #"+str(i+1)+":",str(k),str(r),str(c))
```

## 第二题

```python
def right(s):
    end = 0
    for i in range(len(s)-1,-1,-1):
        if s[i] == ")":
            end += 1
        else:
            return end
def remove(s,n):
    return s[n:-n]

nums = int(input())
result = [0]*nums
for i in range(nums):
    result[i] = list(input())

for i in range(nums):
    for j in range(len(result[i])):
        result[i][j] = "("*int(result[i][j])+result[i][j]+")"*int(result[i][j])
    while len(result[i])>1:
        if result[i][0][-1] == '0' or result[i][1] == '0':
            result[i][0] += result[i][1]
            result[i].pop(1)
        else:
            mind = min(right(result[i][0]),right(result[i][1]))
            result[i][1] =remove(result[i][1],mind)
            result[i][0] = result[i][0][:-mind]+result[i][1]+result[i][0][-mind:]
            result[i].pop(1)
    print("Case #"+str(i+1)+":",result[i][0])
            
result
```

我的想法是这样，分三种情况：
第一种情况：如果 K 能被 N 整除，最简单的情况，生成[1...N]的列表，对列表进行右移(0~N-1)位的操作并存储，再根据对角线的 index 查找相同 index 相同数字的列表，最后输出。
如果 K 不能被 A 整除，找出尽量多的（N-2个）A，再找出和 A 不相等的 X 和 Y，使得 X+Y+(N-2)$\times$A=K，分两种情况：
存在 X,Y ,X!=Y,假设 Y 在 test[N-1][N-1], X 在 test[N-2][N-2],生成一个 [A,X,[1...N except for A,X,Y],Y] 的列表，之后的操作同第一种情况
第三种，最麻烦的一种，如果不存在 X,Y 使得 X!=Y,生成一个 [A,X[1...N except for A,X,]] 的列表，从(0~N-4)行的操作同上，
从 N-3 行开始，N-3 行第0列数字写上 X(因为没其他地方写了)，(1~N-4) 列随机选取 set(col[N-2]) & set(col[N-1]) 的数，N-3 列填 A，N-2 列填 （全集差集（N-2 列并集 N-3行）），N-1 列填 全集差集 N-3 行。
接下来只剩(N-2)$\times$2 的方格需要填（因为[N-2][N-2]，[N-2][N-1],[N-1][N-2],[N-1][N-1] 都可以确定），对 N-2 行，i 列填（全集差集（i 列并集 N-2行））如果第 i 列求得的差集为空集，找全集与 N-2 行的差集记为 remain，在(0~i-1)列集合中查找并集 remain 为空集的,记为列 j，将 [N-2][j] 的值与 remain 交换，如此循环，直至 remain 并集 i 列为空集。
最后 N-1 行的 i 列只需（全集差集 i 列）即可。
